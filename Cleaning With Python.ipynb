```python
age_categories = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=age_categories)
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Every 3 months': 90
}

df['purchase_frequency'] = df['frequency_of_purchases'].map(frequency_mapping)
df['purchase_frequency']
```

```python
# Check which values weren't mapped and their counts
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Every 3 months': 90
}

unique_vals = set(df['frequency_of_purchases'].dropna().unique())
unmapped = sorted(list(unique_vals - set(frequency_mapping.keys())))
print('Unmapped values:', unmapped)
print('\nCounts of all values:')
print(df['frequency_of_purchases'].value_counts())
```
